```python
subset_df = df
    
sum_by_area = subset_df.groupby('Area')[ [ 'Y1985', 'Y1986', 'Y1987',
       'Y1988', 'Y1989', 'Y1990', 'Y1991', 'Y1992', 'Y1993', 'Y1994', 'Y1995',
       'Y1996', 'Y1997', 'Y1998', 'Y1999', 'Y2000', 'Y2001', 'Y2002', 'Y2003',
       'Y2004', 'Y2005', 'Y2006', 'Y2007', 'Y2008', 'Y2009', 'Y2010', 'Y2011',
       'Y2012', 'Y2013', 'Y2014', 'Y2015']].sum()
    
sum_by_area_df = pd.DataFrame(sum_by_area)
```

```python
sum_by_area_df
```

```python
# Assuming your DataFrame is named df
# You can adjust the id_vars, value_vars, and var_name as needed
sum_by_area_df = sum_by_area_df.reset_index()
# Specify the columns to use as identifiers (id_vars) and the columns to melt (value_vars)
id_vars = ['Area']
value_vars = sum_by_area_df.columns[1:]  # Exclude the first two columns ('Area Code' and 'Area')

# Use the melt function to reshape the DataFrame from wide to long format
df_long = pd.melt(sum_by_area_df, id_vars=id_vars, value_vars=value_vars, var_name='Year', value_name='Value')

# Display the resulting long-format DataFrame
print(df_long)
```
